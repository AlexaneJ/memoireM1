```python
import json
import glob
import re
import codecs

print(glob.glob("*"))

def ouvrir_fichier(path):
    f = codecs.open(path, encoding="utf-8")
    chaine = f.read()
    f.close()
    return chaine

def decouper_mots(texte):
    mots = re.findall("[a-zA-ZÀ-ÿ][a-zA-ZÀ-ÿ-]*", texte)
    return mots

def write_json(path, contenu):
    w = codecs.open(path, "w", encoding="utf-8")
    w.write(json.dumps(contenu, indent = 2, ensure_ascii=False))
    w.close()
    
def ouvrir_json(path):
    f = codecs.open(path, encoding="utf-8")
    toto = json.load(f)
    f.close()
    return toto
```

```python
folder1 = "anno_unification_spacy_v2/"
folder2 = "annoMan_unificationSpacy/"
```

```python
liste_fic = ["APIL", "Wikinews", "Sequoia", "aijwikiner", "GSD", "spoken", "pg6470"]
```

```python
ext1 = "_spacy.json"
ext2 = "_pourSpacy.json"
```

```python
#on créé un dico des annotations CasEN
dicoSpaCy = {}
for i in liste_fic:
    path_fic = folder1+i+ext1
    fic = ouvrir_json(path_fic)
    for j in fic:
        dicoSpaCy[j.strip()]=(fic[j]).strip()
```

```python
def compare(first, second):
    commun = {}
    sharedKeys = set(first.keys()).intersection(second.keys())
    print("D'abord, toutes les clés identiques avec valeurs identiques : ")
    for key in sharedKeys:
        if first[key]==second[key]:
            print("Key: {}, Value 1: {}, Value 2: {}".format(key, first[key], second[key]))
            commun[key] = ("Value 1: {}, Value 2: {}".format(first[key], second[key]))
    print("-------------------------")
    print("Puis les clés identiques dont les valeurs ne sont pas identiques :")
    for key in sharedKeys:
        if first[key] != second[key]:
            print("Key: {}, Value 1: {}, Value 2: {}".format(key, first[key], second[key]))
    diffKeys = set(first.keys()).difference(second.keys())
    print("-------------------------")
    print("Puis tout ce qui est différent :")
    for key in diffKeys:
        if key in first : 
            print("Key 1: {}, Value 1 : {}".format(key, first[key]))
        if key in second:
            print("Key 2: {} , Value 2: {}".format(key, second[key]))
    write_json("corpusEntier_exactitudeTotale_SpaCyvsMano.json", commun)
    
    return (commun, len(commun))
    
compare(dicoSpaCy, dicoMano)
```
