```python
def ouvrir_fichier(path):
    f = codecs.open(path, encoding="utf-8")
    chaine = f.read()
    f.close()
    return chaine

def decouper_mots(texte):
    mots = re.findall("[a-zA-ZÀ-ÿ][a-zA-ZÀ-ÿ-]*", texte)
    return mots

def write_json(path, contenu):
    w = codecs.open(path, "w", encoding="utf-8")
    w.write(json.dumps(contenu, indent = 2, ensure_ascii=False))
    w.close()
    
def ouvrir_json(path):
    f = codecs.open(path, encoding="utf-8")
    toto = json.load(f)
    f.close()
    return toto
```

```python
liste_fic = ["aijwikiner", "APIL", "GSD", "pg6470", "Sequoia", "spoken", "Wikinews"]
```

```python
spacy1="annotations_spacy_v3_nonOriginal/"
spacy2 ="_anno_spacy.txt"
```

```python
def get_anno_spacy(name_fic, path_fic):
    f = open(path_fic, encoding="utf-8")
    dico_spacy = {}
    for ligne in f:
        elems = re.split("\s\s+", re.sub("\s\s+", "|", ligne))
        for i in elems:
            elem=i.split("|")
            forme = elem[0]
            #print(forme)
            start = elem[1]
            #print(start)
            end = elem[2]
            #print(end)
            anno = elem[3]
            #print(anno)
            anno = re.sub("\\n", "", anno)
            if elem[0] != "S":
                dico_spacy[forme] = [start, end, anno]
                
    print(len(dico_spacy))
    w = open("%s_spacy.json"%name_fic, "w")
    w.write(json.dumps(dico_spacy))
    w.close()
    return ouvrir_json("%s_spacy.json"%name_fic)
```

```python
for i in liste_fic:
    path_fic = spacy1+i+spacy2
    name_fic = i
    get_anno_spacy(name_fic, path_fic)
```
