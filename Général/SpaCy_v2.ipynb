```python
nlp2 = spacy.load("fr_core_news_md")
```

```python
def ecrire_fichier(chemin, liste):
    w = open(chemin, "w", encoding="utf-8")
    for i in liste:
        w.write("%s \t %s\n"%i)
    w.close()
```

```python
def trouve_entite(doc, nom_fichier):
    liste = []
    for ent in doc.ents :
        contenu = (ent.text, ent.label_)
        liste.append(contenu)
    ecrire_fichier("annotations_spacy_v2/%s_anno_spacy.txt" %nom_fichier, liste)
```

# Essai sur CorpusAPIL.txt

```python
nom_fichier1 = "annotations_spacy_v2/%s"
my_file = open(nom_fichier1%"CorpusAPIL_test.sample.txt", "r")
text = my_file.read()
```

```python
# on process le texte
doc = nlp2(text)

for token in doc:
    # Get the token text, part-of-speech tag and dependency label
    token_text = token.text
    token_pos = token.pos_
    token_dep = token.dep_
    # This is for formatting only
    print(f"{token_text:<12}{token_pos:<10}{token_dep:<10}")
```

```python
for ent in doc.ents:
    print(ent.text, ent.start_char, ent.end_char, ent.label_)
```

```python
trouve_entite(doc, "APIL_result")
```
